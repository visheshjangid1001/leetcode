```python
def MaximumSubarraySum():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    maxi = float("-inf")
    for i in range(0,n):
        total = 0
        for j in range(i,n):
            total = total + nums[j]
            maxi = max(maxi,total)
    return maxi

result = MaximumSubarraySum()
print(result)
```

```python
def MaximumSubarraySum():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    maxi = float('-inf')
    total = 0
    for i in range(0,n):
        total = total + nums[i]
        maxi = max(maxi,total)
        if total<0:
            total = 0
    return maxi
            
    
    
result = MaximumSubarraySum()
print(result)
```
