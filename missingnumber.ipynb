## Brute force solution

```python

def find_missing_number():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    for i in range(0,n+1):
        if i not in nums:
            return i

missing = find_missing_number()
print(missing)
```

## Better solution
### using dictionary

```python
def find_missing_number():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    freq={}
    for i in range(0,n+1):
        freq[i] = 0
    for num in nums:
        freq[num] = 1
    for k,v in freq.items():
        if v == 0:
            return k

missing = find_missing_number()
print(missing)
```

### Optimal Solution

```python
def find_missing_number():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    return (n * (n + 1)) // 2 - sum(nums)
    

missing = find_missing_number()
print(missing)
```
