```python
def MaxConsecutiveOnes():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    count = 0
    max_count = 0 
    for i in range(0,len(nums)):
        if nums[i] == 1:
            count += 1
        else:
            max_count = max(max_count,count)
            count=0
    return max(max_count,count)

Max = MaxConsecutiveOnes()
print(Max)
```
