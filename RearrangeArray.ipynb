```python
def RearrangeAray():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    pos = []
    neg = []
    for num in nums:
        if num > 0:
            pos.append(num)
        else:
            neg.append(num)
    for i in range(0,len(pos)):
        nums[2*i] = pos[i]
        nums[(2*i)+1] = neg[i]
    return nums
result = RearrangeAray()
print(result)
```

```python
def RearrangeAray():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    result = [0]*n
    P,N = 0,1
    for i in range (0,n):
        if nums[i] >= 0:
            result[P] = nums[i]
            P+=2
        else:
            result[N] = nums[i]
            N+=2
    return result
    
    
result = RearrangeAray()
print(result)
```
