Binary Search is an efficient searching algorithm used to find an element in a sorted array.

🔹 Basic Idea

Instead of checking every element (like linear search), binary search:

👉 Repeatedly divides the search space into half

🔹 How it Works (Step-by-Step)

Suppose array is:

[1, 3, 5, 7, 9, 11]

Find 7

Find middle element
→ middle = 5

Compare:

7 > 5 → search in right half

New array:

[7, 9, 11]

Again find middle → 9

Compare:

7 < 9 → search in left half

Now:

[7]

```python
def binarysearch(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    while low <= high:
        mid = (low+high)//2
        if nums[mid] == target:
            return mid 
        elif nums[mid] > target:
            high = mid - 1
        else:
            low = mid+1
    return -1

nums = list(map(int, input("Enter numbers separated by space: ").split()))
target = int(input("Enter target value: "))

result = binarysearch(nums,target)
print(result)
```
