### Find Minimum In Rotated Sorted Array 

```python
def FindMinimum():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    mini = float("inf")
    for i in range(n):
        mini = min(mini,nums[i])
    return mini
```

```python
def FindMinimum():
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    n = len(nums)
    low = 0
    high = n-1
    mini = float("inf")
    while low<=high:
        mid = (low+high)//2
        if nums[mid]<=nums[high]:
            mini = min(mini,nums[mid])
            high = mid-1
        else:
            mini = min(mini,nums[low])
            low = mid+1
    return mini

result = FindMinimum()
print(result)
        
```
