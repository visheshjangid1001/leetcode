```python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None


class Solution:
    def removeElements(self, head, val):
        dummy = ListNode(0)     # fixed
        dummy.next = head       # important
        prev = dummy
        curr = head

        while curr is not None:
            nxt = curr.next
            if curr.val == val:
                prev.next = nxt
            else:
                prev = curr
            curr = nxt

        return dummy.next       # fixed


# -------- Helper Function --------
def print_list(head):
    temp = head
    while temp:
        print(temp.val, end=" -> ")
        temp = temp.next
    print("None")


# -------- Create Linked List --------
# Example: 1 -> 2 -> 6 -> 3 -> 4 -> 5 -> 6
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(6)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(4)
head.next.next.next.next.next = ListNode(5)
head.next.next.next.next.next.next = ListNode(6)


# -------- Run Solution --------
sol = Solution()
new_head = sol.removeElements(head, 6)

# -------- Output --------
print("Updated List:")
print_list(new_head)
```
