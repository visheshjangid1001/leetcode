### Find the Length of loop

#### Brute force solution

```python
class Solution:
    def findlength(self, head):
        temp = head
        travel = 0
        my_dict = dict()

        while temp is not None:
            if temp in my_dict:
                return travel - my_dict[temp]   # length of cycle
            
            my_dict[temp] = travel
            travel += 1
            temp = temp.next

        return 0   # no cycle found
```

#### Optimal solution

```python
class Solution:
    def findlength(self, head):
        slow = head
        fast = head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                slow = slow.next
                count = 1
                while slow!= fast:
                    slow = slow.next
                    count+=1
                return count
        return 0
```

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None


class Solution:
    def findlength(self, head):
        slow = head
        fast = head

        # Step 1: Detect cycle
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                # Step 2: Find cycle length
                count = 1
                slow = slow.next

                while slow != fast:
                    slow = slow.next
                    count += 1

                return count

        return 0   # No cycle


# ------------------ DRIVER CODE ------------------

# Create nodes
head = ListNode(1)
second = ListNode(2)
third = ListNode(3)
fourth = ListNode(4)
fifth = ListNode(5)
sixth = ListNode(6)
seventh = ListNode(7)

# Connect nodes
head.next = second
second.next = third
third.next = fourth
fourth.next = fifth
fifth.next = sixth
sixth.next = seventh

# Create cycle (7 -> 3)
seventh.next = third

# Run function
sol = Solution()
result = sol.findlength(head)

print("Cycle Length:", result)
```
