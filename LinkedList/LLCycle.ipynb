```python
def cycle(self):
    slow = self.head
    fast = self.head
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Create cycle (for testing)
    def create_cycle(self, pos):
        if pos == -1:
            return
        
        cycle_node = None
        temp = self.head
        index = 0
        
        while temp.next:
            if index == pos:
                cycle_node = temp
            temp = temp.next
            index += 1
        
        temp.next = cycle_node  # create cycle

    # 🔥 Detect cycle
    def cycle(self):
        slow = self.head
        fast = self.head
        
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            
            if slow == fast:
                return True
        
        return False


# -------- Main Program --------
ll = LinkedList()

n = int(input("Enter number of elements: "))

for i in range(n):
    val = int(input(f"Enter element {i+1}: "))
    ll.insert(val)

# Create cycle
pos = int(input("Enter position to create cycle (-1 for no cycle): "))
ll.create_cycle(pos)

# Check cycle
if ll.cycle():
    print("Cycle detected in linked list")
else:
    print("No cycle in linked list")
```
