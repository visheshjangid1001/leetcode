## What is a Linked List?

A Linked List is a linear data structure where elements are connected using pointers instead of being stored in continuous memory like arrays.

👉 Think of it like a train 🚆

Each coach = node
Each coach is connected to the next one
## Structure of Linked List

Each element is called a Node, and it has 2 parts:

[ DATA | NEXT ]
DATA → actual value (e.g., 10, 20)
NEXT → address of next node
🔸 Example
10 → 20 → 30 → 40 → NULL

👉 Internally:

[10 | address_of_20] → [20 | address_of_30] → [30 | address_of_40] → [40 | NULL]
First node = HEAD
Last node points to NULL
## How It Works (Important)

Unlike arrays:

Array: stored in continuous memory
Linked List: stored anywhere in memory, connected by pointers

👉 That’s why:

No index like arr[2]
You must traverse from head
## Types of Linked List

1️⃣ Singly Linked List
10 → 20 → 30 → NULL
Only forward direction
Each node has only next


2️⃣ Doubly Linked List
NULL ← 10 ⇄ 20 ⇄ 30 → NULL
Structure:
[Prev | Data | Next]
👉 Advantages:
Traverse forward & backward


3️⃣ Circular Linked List
10 → 20 → 30 → (back to 10)
👉 Last node points to head instead of NULL

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
        
```

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
    ## append val in linked list
    def append(self,val):
        new_node = Node(val)
        if self.head == None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
            
    ## Traverse 
    def traverse(self):
        if self.head == None:
            print("sll is empty")
        else:
            current = self.head
            while current is not None:
                print(current.val,end=" ")
                current = current.next
                
                
sll = SinglyLinkedList()
sll.append(2)
sll.append(6)
sll.append(10)
sll.append(9)
sll.append(8)
sll.traverse()
```
