```python
rows = int(input("Enter rows: "))
cols = int(input("Enter columns: "))

nums = []

for i in range(rows):
    row = list(map(int, input().split()))
    nums.append(row)

for i in range(0, rows):
    for j in range(0, cols):
        print(nums[i][j], end=" ")
    print()
```

```python
rows = int(input("Enter rows: "))
cols = int(input("Enter columns: "))

nums = []

for i in range(rows):
    row = list(map(int, input().split()))
    nums.append(row)
    
for i in range(0,rows):
    for j in range(0,cols):
        if i + j == cols - 1:
            print(nums[i][j],end=" ")
        else:
            print("*",end=" ")
    print()
```
