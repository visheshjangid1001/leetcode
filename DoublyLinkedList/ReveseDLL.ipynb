## Reverse Doubly Linked List

#### Brute force solution

```python
def Reverse(self,head):
    temp = head
    while temp is not None:
        stack.append(temp.val)
        temp = temp.next
    temp = head
    while temp is not None:
        e = stack.pop()
        temp.val = e
        temp = temp.next
    return head
```

#### Optimal solution

```python
def Reverse(self,head):
    if head.next is None:
        return head
    cur = head
    prev = None
    while cur is not None:
        front = cur.next
        cur.next = prev
        cur.prev = front
        prev = cur
        cur = front
    return prev
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class Solution:
    def reverseDLL(self, head):
        if head is None:
            return None

        cur = head
        prev = None

        while cur is not None:
            front = cur.next

            # reverse links
            cur.next = prev
            cur.prev = front

            prev = cur
            cur = front

        return prev   # new head


# -------- Helper Functions --------
def print_forward(head):
    temp = head
    while temp:
        print(temp.data, end=" <-> ")
        temp = temp.next
    print("None")


def print_backward(head):
    temp = head
    while temp.next:
        temp = temp.next   # go to last node
    
    while temp:
        print(temp.data, end=" <-> ")
        temp = temp.prev
    print("None")


# -------- Create DLL --------
head = Node(10)
second = Node(20)
third = Node(30)
fourth = Node(40)

# Linking
head.next = second
second.prev = head

second.next = third
third.prev = second

third.next = fourth
fourth.prev = third


# -------- Before Reverse --------
print("Original List (Forward):")
print_forward(head)


# -------- Reverse --------
sol = Solution()
new_head = sol.reverseDLL(head)


# -------- After Reverse --------
print("\nReversed List (Forward):")
print_forward(new_head)

print("\nReversed List (Backward):")
print_backward(new_head)
```
